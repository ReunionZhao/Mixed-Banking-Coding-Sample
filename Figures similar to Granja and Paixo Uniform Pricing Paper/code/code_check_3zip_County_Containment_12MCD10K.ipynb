```python
import pandas as pd

df = pd.read_csv('/Users/zhaorunping/Desktop/Research_Onging/2502_Haas_Matteo/Mixed-Banking/Figures similar to Granja and Paixo Uniform Pricing Paper/data/intermediate/for_reg_data_multi_branch_sample_932466/12MCD10K_data.csv')
df.head()
```

```python
# Calculate the number of unique zip_3digit values within each county
county_zip_counts = df.groupby('county')['zip_3digit'].transform('nunique')

# Add a new column to the original df to indicate the number of unique zip_3digit values in each county
df['zip_3digit_count_in_county'] = county_zip_counts

# Calculate the number of unique counties within each zip_3digit
zip_county_counts = df.groupby('zip_3digit')['county'].transform('nunique')

# Add a new column to the original df to indicate the number of unique counties in each zip_3digit
df['county_count_in_zip_3digit'] = zip_county_counts
```
